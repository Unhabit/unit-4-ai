# Shakespeare GPT

```python
session2= gpt2.start_tf_sess()
```

```python
gpt2.finetune(
    session2,
    'shakespeare.txt',
    model_name='124M',
    steps=100,
    run_name='shakespeare'
)
```

```python
gpt2.generate(
    session2,
    prefix='Shovo : He will not survive ',
    run_name='shakespeare'
)
```
